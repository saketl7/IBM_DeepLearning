```
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

## Using Categorical Encoding i.e. LabelEncoder()

```
le = LabelEncoder()
le = le.fit_transform(df.iloc[:,-1])
#label.shape <- (150,) reason

le = le.reshape((-1,1))
le.shape
```

```
df.iloc[:,-1] = le
df
```

## Checking for NaN values

```
df.isna().sum()
```

## Using Scaling technique [StandardScaler()] to normalize the dataset

```
scaler = StandardScaler()
input_float = df.iloc[:,0:4]
input_float = scaler.fit_transform(input_float)
```

```
input_float
```

### Replacing the normalized values in the dataset

```
df.iloc[:,0:4] = input_float

df.head()
```

### Making 3 new columns for 3 different flower types

```
df.insert(5, "flower1", 0)
df.insert(6, "flower2", 0)
df.insert(7, "flower3", 0)
df
```

```
[rows, columns] = df.shape

for i in range(rows):
    if df.iloc[i,4] == 0:
        df.iloc[i,5] = 1
    elif df.iloc[i,4] == 1:
        df.iloc[i,6] = 1
    else:
        df.iloc[i,7] = 1

df
```
